```python
import numpy as np

# Define the function f(x) = x^2
def f(x):
    return x**2

# Define the point at which to approximate the gradient
x0 = -1.0

# Define the step size for the limit approximation
h = 0.0001

# Approximate the gradient using the limit definition
grad = (f(x0 + h) - f(x0)) / h

print(f"Gradient at x={x0}: {grad}")
```

```python
# plot the function and the gradient as a tangent line
import matplotlib.pyplot as plt

# Define the x range for the plot
x = np.linspace(-6, 6, 100)

# Plot the function
plt.plot(x, f(x))

# Plot the gradient as a tangent line
plt.plot(x0, f(x0), 'ro')
plt.plot(x0+h, f(x0+h), 'ro')
plt.plot(x, grad*(x - x0) + f(x0), 'r--')

# Add labels and legend
plt.xlabel('x')
plt.ylabel(r'$f(x)=x^2$')
plt.legend([r'$f(x)=x^2$', f'x={x0}, x+h={x0+h}'])
plt.title(rf'Difference Quotient for $f(x)=x^2$, $h={h}$')

# set x and y limits
plt.xlim(-6, 6)
plt.ylim(-10, 30)


# Show the plot
plt.show()
```
